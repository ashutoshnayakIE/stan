<b>data types</b>

<ul>
    <li>real a : real values a</li>
    <li>int a  : integer values a</li>
    <li>real[10] a : real values an array a of size 10</li>
    <li>real a : real values a</li>
    <li>vector[size] a : is a column vector</li>
    <li>row_vector[size] a : is a row vector</li>
    <li>matrix[r,c] a : is a matrix</li>
    <li>vector[size] a : is a column vector</li>
    <li>simplex[size1] a[size2] : a is a simplex of size size1 and each element of a is itself a simplex of size size2</li>
</ul>

The following are the constrained data types:
<ul>
    <li>simplex[K] a : sum of all the elements in a = 1</li>
    <li>unit_vector[K] x : the value of x will be constrained to be a vector of size K with unit length</li>
    <li>ordered a[10] : ordered vector of size 10</li>
    <li>positive_ordered a[10] : positively ordered vector size 10 (elements are constrained to be positive)</li>
</ul>

```python
import pandas as pd
import numpy as np
import pystan
import arviz
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
np.random.seed(42)
Y1 = np.random.normal(97,15,30)
Y2 = np.random.normal(108,15,30)
```

The following block shows an example as how the data is entered, how to define parameters and how to build the bayesian model. We provide a prior on mu and sigma.

If no prior is provided, stan uses non-informative weak priors such as uniform.

stan uses hamiltonian MCMC (and NUTS algorithm for drawing samples)

```python
two_group = """
data {
    int nY1 ;        # initialize a variable to indicate the number of elements in Y
    int nY2 ;        # initialize a variable to indicate the number of elements in Y
    vector[nY1] Y1 ; # initialize a vector to hold the observations
    vector[nY2] Y2 ; # initialize a vector to hold the observations
}
parameters {
    real mu1 ;            #mean of the Y1 data-generating population
    real mu2 ;            #mean of the Y2 data-generating population
    real<lower=0> sigma ; #SD of both data-generating populations, constrained to positive
}
model {
    mu1 ~ normal(100,20) ;   #prior on mu1
    mu2 ~ normal(100,20) ;   #prior on mu2
    sigma ~ cauchy(0,15) ;   #prior on sigma
    Y1 ~ normal(mu1,sigma) ; #sampling statement for Y1 data
    Y2 ~ normal(mu2,sigma) ; #sampling statement for Y2 data
}
"""
```

```python
# data needs to be entered in a dictionary format
data = {'nY1': 30,
        'nY2': 30,
        'Y1': Y1,
        'Y2': Y2}
```

```python
model = pystan.StanModel(model_code=two_group)
fit   = model.sampling(data=data, iter=10000,warmup=5000, refresh=1000, chains=4)
```

```python
fit
```

```python
fit.plot()
```

<h1>Example on weather</h1>

```python
import matplotlib.pyplot as mplt
```

```python
data = pd.read_csv('hmm_example.csv')
y = data['y']
z = data['z']

fig = mplt.figure(figsize= (18,4))
mplt.subplot(1,2,1)
mplt.plot(z)
mplt.subplot(1,2,2)
mplt.plot(y)
```

```python
# HMM example data (hmm_example.RDS)

# transition matrix (2 states)
theta  = np.array([[0.8,0.2],[0.1,0.9]])

# emmission vector (2 states)
psi    = np.array([3,9])

# timesteps
T = 100

I = 100

# hidden states
Z = np.zeros((I,T),dtype=int)

for i in range(I):
    Z[i,0] = np.random.choice(2,1,p=[0.9,0.1])[0]
    for t in range(2,T):
        s = np.random.choice(2, 1, p = theta[Z[i,t-1]])[0]
        Z[i,t] = s
    
# observations
Y = np.zeros((I,T))
for i in range(I):
    for t in range(T):
        Y[i,t]  = np.random.normal(psi[Z[i,t]], 1, 1)[0]

# visualization
fig = mplt.figure(figsize= (18,4))
mplt.subplot(1,2,1)
mplt.plot(Z[0])
mplt.subplot(1,2,2)
mplt.plot(Y[0])
```

```python
model = """
data {
  int<lower=0> I;
  int<lower=0> T;
  int<lower=0> N;
  real y[I,T];
}

parameters {
  simplex[N] ppi;
  simplex[N] theta[N];
  ordered[N] mu;
}


model {

  // prior
  
  for (s in 1:N){
      target+= normal_lpdf(mu[s] |5,5);
  }
  ppi ~ dirichlet(rep_vector(1,N)) ;
    
  // forward algorithm
  {
      for (i in 1:I){
          real acc[N];
          real gamma[T, N];

          for (k in 1:N){
            gamma[1, k] = log(ppi[k]) + normal_lpdf(y[1] | mu[k], 1);
            //gamma[1, k] = normal_lpdf(y[1] | mu[k], 1);
          }

          for (t in 2:T) {
              for (k in 1:N) {
                  for (j in 1:N)
                      acc[j] = gamma[t-1, j] + log(theta[j, k]) + normal_lpdf(y[i,t] | mu[k], 1);
                  gamma[t, k] = log_sum_exp(acc);
              }
          }
          target += log_sum_exp(gamma[T]);
       }
   }
}
      

"""
```

```python
non_centered = """
data {
  int<lower=0> I;
  int<lower=0> T;
  int<lower=0> N;
  real y[I,T];
}

parameters {
  simplex[N] ppi;
  simplex[N] theta[N];
  real mu;
  real<lower=0> tau;
  real theta_tilde[N];
}

transformed parameters {
  real mean_[N];
  for (k in 1:N)
    mean_[k] = mu + tau * theta_tilde[k];
}

model {

  // prior
  
  mu          ~ normal(5, 5);
  tau         ~ cauchy(0, 2);
  theta_tilde ~ normal(0, 1);
  ppi         ~ dirichlet(rep_vector(1,N)) ;
    
  // forward algorithm
  {
      for (i in 1:I){
          real acc[N];
          real gamma[T, N];

          for (k in 1:N){
            gamma[1, k] = log(ppi[k]) + normal_lpdf(y[1] | mean_[k], 1);
          }

          for (t in 2:T) {
              for (k in 1:N) {
                  for (j in 1:N)
                      acc[j] = gamma[t-1, j] + log(theta[j, k]) + normal_lpdf(y[i,t] | mean_[k], 1);
                  gamma[t, k] = log_sum_exp(acc);
              }
          }
          target += log_sum_exp(gamma[T]);
       }
   }
}
      

"""
```

```python
modelnonc = pystan.StanModel(model_code = non_centered)
```

```python
n = 20
stan_data  = {"I": n,
              "T": 100,
              "N": 2,
              "y": Y[0:n]}
fitnonc = modelnonc.sampling(data=stan_data, iter=100, seed = 0, chains=4, control=dict(adapt_delta=0.8, max_treedepth=10))
```

<h1>Disadvantages with bayesian implementation of HMM </h1>

<ul>
    <li>Multi-modality may lead to non-mixing of the chains in stan NUTS sampler. </li>
    <li>HMMs can be hard to optimize using EM algorithm as the HMM model becomes more complicated, e.g. non-homogenous HMM or non-stationary HMM. For these problems, we need to consider MCMC techniques to obtain the model parameters.</li>
    <li>HMM performance depends on the priors selected. As more and more data is collected, likelihood depends more on the posterior. Reducing the amount of data, however, reduces the influence of the likelihood and makes it easier to corral the mixture components.</li>
    <li>HMM may not be well suited for large data sets due to the fact that HMMs are slow (as told by Geoffrey Hinton), as the data sets increase in size, it cant fit well as  -- When you add more data your posterior becomes more concentrated, and as it concentrates it becomes more sensitive to the compatibility of your model and the true data generating process. In other words, the more data you have the more your posterior manifests model misfit. Often this model misfit manifests in awkward posterior geometries that make computation difficult, yielding behavior not unlike what you see. (https://discourse.mc-stan.org/t/chains-stuck-when-use-larger-dataset-but-not-smaller/1088) </li>
</ul>
